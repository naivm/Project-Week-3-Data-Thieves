# TO GET DESTINATIONS

```python
#to get destinations
'''
1) only countries within the gini table
2) filter by year
3) 
'''

countries_lst = list(gini['Country'])
#print(countries_lst)
```

```python
print(len(countries_lst))
```

```python
countries_from_migrations = list(migrations['Area/Region/Country'])
#print(countries_from_migrations)

countries_for_mig_table = []

for country in countries_from_migrations:
    if country in countries_lst and country not in countries_for_mig_table:
        countries_for_mig_table.append(country)

print(len(countries_for_mig_table))
#print(countries_for_mig_table)
```

```python
countries_missing = [country for country in countries_lst if country not in countries_from_migrations]
print(countries_missing)
print(len(countries_missing))
```

```python
migrations_destinations = migrations[migrations['Area/Region/Country'].isin(countries_for_mig_table)]

migrations_destinations.shape
```

```python
migrations_dest_2015 = migrations_destinations[migrations_destinations['Year'] == 2015]
migrations_dest_2015.shape
```

```python
migrations_dest_2017 = migrations_destinations[migrations_destinations['Year'] == 2017]
migrations_dest_2017.shape
```

```python
#rename countries columns on all tables
migrations_dest_2015 = migrations_dest_2015.rename(columns={'Area/Region/Country':'Country','Country or Area of Origin / Total':'Total 2015'})
migrations_dest_2017 = migrations_dest_2017.rename(columns={'Area/Region/Country':'Country','Country or Area of Origin / Total':'Total 2017'})
```

```python
migrations_dest_2015.head()
```

```python
#merge the two migrations tables
migrations_2015_2017 = pd.merge(migrations_dest_2015, migrations_dest_2017, on = 'Country')
migrations_2015_2017.head()
```

```python
#merge gini and migrations tables
gini_migrations = pd.merge(gini, migrations_2015_2017, on = 'Country')
gini_migrations.head()
```

```python
gini_migrations = gini_migrations.drop(['Year_x','Year_y'], axis =1)
```

```python
gini_migrations.head()
```

```python
#gini_migrations.describe()
```

```python
gini_migrations["Balance (%)"] = ((gini_migrations['Total 2017'] - gini_migrations['Total 2015']) / gini_migrations['Total 2017']) * 100
```

```python
#gini_migrations.head()
```

```python
#
gini_migrations['Balance (%)'] = gini_migrations['Balance (%)'].map("{0:.1f}".format)
```

```python
gini_migrations['Balance (%)'] = gini_migrations['Balance (%)'].astype('float')
```
