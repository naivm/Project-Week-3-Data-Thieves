# Binning of Gini-Index

```python
happy_gini.isnull().sum()
```

```python
happy_gini["gini"] = happy_gini['gini'].fillna(0)
```

```python
# Binning of Gini-Index

gini = ["no Data","low", "middle", "high"]

cutoffs = ["0","24", "37","50","65"]

happy_gini["grade_gini"] = pd.cut(happy_gini["gini"],cutoffs,labels = gini)



happy_gini.head()
```

```python
#(str(happy_gini.loc[2,("gini")])).replace("0.0","No Data")
happy_gini_prepare = happy_gini.replace(0.0, np.nan)
```

```python
#happy_gini["gini"] = happy_gini["gini"].isnull()#.fillna("No Data")
happy_gini_prepare["gini"] = happy_gini_prepare["gini"].fillna("no Data")
```

```python
happy_gini_cut =happy_gini_prepare[["Country","Happiness.Rank","Happiness.Score","gini","grade_gini"]].sort_values(by = "Happiness.Rank", ascending= True)
```

```python
happy_gini_cut_rename = happy_gini_cut.rename(columns={"Happiness.Rank": "Happiness Rank","Happiness.Score":"Happiness Score"})

```

```python
happy_gini_cut_rename.set_index("Happiness Rank")
```
