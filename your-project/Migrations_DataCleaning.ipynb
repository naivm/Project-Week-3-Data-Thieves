```python
#drop NaN columns

#create list of columns to be droped
col_drop = ['Unnamed: ' + str(i) for i in range(242,16324)]
#print(col_drop)

#drop the columns
migrations = migrations.drop(col_drop, axis =1)

migrations.head()
```
