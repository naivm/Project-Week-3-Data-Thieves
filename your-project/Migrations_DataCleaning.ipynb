```python
# import libraries
import pandas as pd
import numpy as np
```

```python
#import csv file and create data frame
migrations = pd.read_csv('UN_MigrantStockByOriginAndDestination_2017.csv',low_memory=False)
migrations.head(50)
```

```python
#check for NaN values 
migrations.isnull().sum()
migra_null = migrations.isnull().sum()
migra_null [migra_null > 0]
```

```python
#check columns names
migrations.columns
```

```python
#drop NaN columns

#create list of columns to be droped
col_drop = ['Unnamed: ' + str(i) for i in range(242,16324)]
#print(col_drop)

#drop the columns
migrations = migrations.drop(col_drop, axis =1)

migrations.head()
```

```python
#drop top rows, unusefull information

#create list of rows to drop
drop_rows = list(range(13))
#print(drop_rows)

#drop the rows
migrations = migrations.drop(drop_rows, axis = 0)

migrations.head()
```

```python
#drop all rows with only NaN values
migra_no_nan = migrations.dropna(how='all')
migra_no_nan.head()
```

```python
#migrations.shape
migra_no_nan.shape
```

```python
migra_no_nan.tail()
```

```python
#drop unusefull columns
migra_no_nan = migra_no_nan.drop(['Unnamed: 1','Unnamed: 3','Unnamed: 4','Unnamed: 5'], axis = 1)
migra_no_nan.head()
```

```python
#put on headers in the same row and delete row '13'

migra_no_nan.loc[14,'Unnamed: 0'] = 'Year' 
migra_no_nan.loc[14,'Unnamed: 2'] = 'Area/Region/Country' 
migra_no_nan.loc[14,'Unnamed: 6'] = 'Country or Area of Origin / Total' 
migra_no_nan = migra_no_nan.drop(13, axis = 0)

migra_no_nan.head()
```

```python
migra_no_nan.columns
```

```python
#drop last column
migra_no_nan = migra_no_nan.drop('Unnamed: 241', axis = 1)
migra_no_nan.head()
```

```python
#redefine header row

new_header = migra_no_nan.iloc[0] #grab the first row for the header
migra_new_header = migra_no_nan[1:] #take the data less the header row
migra_new_header.columns = new_header #set the header row as the df header

migra_new_header.head()
```

```python
migra_new_header.tail()
```

```python
#delete NaN rows
migra_NaN_rows = migra_new_header.dropna(axis = 0)
migra_NaN_rows.tail()
```

```python
migra_NaN_rows.columns
```

```python
migra_NaN_rows.head()
```

```python
migra_NaN_rows.dtypes
```

```python
#reset indexes
migra_index = migra_NaN_rows.reset_index(drop = True)
```

```python
migra_index.head()
```

```python
#remove commas
migra_comma = migra_index.applymap(lambda x: str(x.replace(',','')))
migra_comma.head()
```

```python
#remove commas
migra_comma = migra_comma.applymap(lambda x: str(x.replace('..','0')))
migra_comma.head()
```

```python
#define data types

#create list of columns
col_list = list(migra_comma.columns)
del col_list[1]
#print(col_list)

for col in col_list:
    migra_comma[col] = migra_comma[col].astype('int64')
    
migra_comma.dtypes
```

```python
migra_comma.tail()
```

```python
migra_from_2014 = migra_comma[migra_comma['Year'] >= 2014]
migra_from_2014.head()
```

```python
migra_from_2014.tail()
```

```python
migra_from_2014.to_csv('migration_2014.csv')
```

```python
migra_from_2014_1 = migra_from_2014[['Year','Area/Region/Country','Country or Area of Origin / Total']]
migra_from_2014_1.head()
```
