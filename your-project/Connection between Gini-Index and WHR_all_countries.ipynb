```python
# List of Countries most happiest and least happiest countries [kaggle: World Happiest Report]

list_of_names = ['Norway', 'Denmark', 'Iceland', 'Switzerland', 'Finland', 'Netherlands','Canada', 'New Zealand', 'Sweden', 'Australia', 'Israel', 'Costa Rica', 'Austria', 'United States of America', 'Ireland', 'Germany', 'Belgium', 'Luxembourg', 'United Kingdom', 'Chile', 'United Arab Emirates', 'Brazil', 'Czech Republic', 'Argentina', 'Mexico', 'Singapore', 'Malta', 'Uruguay', 'Guatemala', 'Panama', 'France', 'Thailand', 'Spain', 'Qatar', 'Colombia', 'Saudi Arabia', 'Trinidad and Tobago', 'Kuwait', 'Slovakia', 'Bahrain', 'Malaysia', 'Nicaragua', 'Ecuador', 'El Salvador', 'Poland', 'Uzbekistan', 'Italy', 'Russia', 'Belize', 'Japan', 'Lithuania', 'Algeria', 'Latvia', 'Moldova', 'Romania', 'Bolivia', 'Turkmenistan', 'Kazakhstan', 'Slovenia', 'Peru', 'Mauritius', 'Cyprus', 'Estonia', 'Belarus', 'Libya', 'Turkey', 'Paraguay', 'Philippines', 'Serbia', 'Jordan', 'Hungary', 'Jamaica', 'Croatia', 'Kosovo', 'China', 'Pakistan', 'Indonesia', 'Venezuela', 'Montenegro', 'Morocco', 'Azerbaijan', 'Dominican Republic', 'Greece', 'Lebanon', 'Portugal', 'Bosnia and Herzegovina', 'Honduras', 'Macedonia', 'Somalia', 'Vietnam', 'Nigeria', 'Tajikistan', 'Bhutan', 'Kyrgyzstan', 'Nepal', 'Mongolia', 'South Africa', 'Tunisia', 'Egypt', 'Bulgaria', 'Sierra Leone', 'Cameroon', 'Iran', 'Albania', 'Bangladesh', 'Namibia', 'Kenya', 'Mozambique', 'Myanmar', 'Senegal', 'Zambia', 'Iraq', 'Gabon', 'Ethiopia', 'Sri Lanka', 'Armenia', 'India', 'Mauritania', 'Mali', 'Ivory Coast', 'Cambodia', 'Sudan', 'Ghana', 'Ukraine', 'Uganda', 'Burkina Faso', 'Niger', 'Malawi', 'Chad', 'Zimbabwe', 'Lesotho', 'Angola', 'Afghanistan', 'Botswana', 'Benin', 'Madagascar', 'Haiti', 'Yemen', 'South Sudan', 'Liberia', 'Guinea', 'Togo', 'Rwanda', 'Syria', 'Tanzania', 'Burundi', 'Central African Republic']

fields = "?fields=alpha3Code;name"
url = "https://restcountries.eu/rest/v2/name/"



contents = [requests.get(url + list_of_names[x] + fields).json() for x in range(len(list_of_names))]

print(contents)
```

```python
d_contents = pd.DataFrame(contents)
d_contents
#[ d_contents[i]  for i in range(len(d_contents))if d_contents[i]=="status"]
```

```python
# Extend Country name with Code (alpha3Code) [Source: API]

country_code =[d_contents[0][i] for i in range(len(d_contents))]
ds = pd.DataFrame.from_dict([country_code[i] for i in range(len(d_contents))])
ds
```

```python
# Merge Data of API (Gini-Index) with Country Name (WH-Report)

codes = list(ds["alpha3Code"])
codes

fields2 = "?fields=gini;name"
url2 = "https://restcountries.eu/rest/v2/alpha/"

ginis = [requests.get(url2 + codes[x] + fields2).content for x in range(len(codes))]

print(ginis)
```

```python
# Transformation

new =[ginis[i] for i in range(len(d_contents))]
new[0]
new1 = [str(new[i]).replace("b'","").replace("'", "") for i in range(len(new))]
print(new1)
new1[0]

import ast

dict_gini = [ast.literal_eval(new1[i]) for i in range(len(new1))]
```

```python
# Connect Gini-Index with Country Name

ds_gini = pd.DataFrame.from_dict([dict_gini[i] for i in range(len(new1))])
ds_gini
```
